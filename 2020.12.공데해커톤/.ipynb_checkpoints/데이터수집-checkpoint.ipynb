```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
```

```python
servicekey = '비밀'

data_url = '비밀2'.format(servicekey)

response = requests.get(data_url)
```

```python
if response.status_code == 200:
    html = response.text
    print(type(html))
    print(html[:150])
    soup = BeautifulSoup(html, 'html.parser')
    
    finded_values = soup.find_all('title')
    print([x.text for x in finded_values][:5])
    
else:
    print("Error Code:" + response.status_code)
```

```python
CommerceInfor = {}

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    
    attr_to_find_list = ['addr1','addr2','title','firstimage','firstimage2','mapx','mapy']
    for each_attr in attr_to_find_list:
        finded_attr = soup.find_all(each_attr)
        print('doing..' + each_attr)
        if CommerceInfor.get(each_attr) is None:
            CommerceInfor[each_attr] = [x.text for x in finded_attr]
        else:
            CommerceInfor[each_attr] = CommerceInfor[each_attr] + [x.text for x in finded_attr]
    
else:
    print("Error Code:" + response.status_code)

print('End!')
# df = pd.DataFrame(CommerceInfor)
print(len(CommerceInfor['addr1']))
print(len(CommerceInfor['addr2']))
print(len(CommerceInfor['title']))
print(len(CommerceInfor['firstimage']))
```
