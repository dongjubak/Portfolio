```python
train_err  = pd.read_csv(PATH+'train_err_data.csv', thousands=',')
train_quality  = pd.read_csv(PATH+'train_quality_data.csv', thousands=',')
train_problem  = pd.read_csv(PATH+'train_problem_data.csv', thousands=',')

test_err  = pd.read_csv(PATH+'test_err_data.csv', thousands=',')
test_quality  = pd.read_csv(PATH+'test_quality_data.csv', thousands=',')
```

# --------- err + quality 합쳐서 돌려보자

```python
# user_id와 errtype만을 사용하여 데이터 셋 생성
# 모든 일자에 대해서 errtype별 발생 건수를 count
# pandas의 groupby를 활용할 경우 큰 연산 자원이 필요.
# numpy로 placeholder를 만들어 구현함.

id_error = train_err[['user_id','errtype']].values
error = np.zeros((15000,42))

for person_idx, err in tqdm(id_error):
    # person_idx - 10000 위치에 person_idx, errtype에 해당하는 error값을 +1
    error[person_idx - 10000,err - 1] += 1
error.shape
```

```python
print(train_err.user_id.max())
print(test_err.user_id.max())
print(train_err.user_id.min())
print(test_err.user_id.min())

print(train_quality.user_id.max())
print(test_quality.user_id.max())
print(train_quality.user_id.min())
print(test_quality.user_id.min())
```

```python
error_df = pd.DataFrame(error)
error_df
```

```python
quality_ex = train_quality.drop_duplicates()
quality_ex = quality_ex.fillna(0)
quality_ex = quality_ex.reset_index(drop=True)
quality_ex = pd.concat([quality_ex['user_id'], quality_ex[quality_ex.columns[3::]]], axis=1)
quality_ex = quality_ex.groupby('user_id')[quality_ex.columns[1::]].sum()
quality_ex = quality_ex.reset_index()
quality_ex
```

```python
ex_df = []
ex_df.append([i for i in range(10000,25000)])

for i in range(13):
    ex_df.append([0 for j in range(15000)])
    
ex_df = pd.DataFrame(ex_df)
ex_df = ex_df.T
ex_df.columns = quality_ex.columns
ex_df = ex_df.set_index(keys='user_id')
ex_df
```

```python
quality_ex = quality_ex.set_index(keys='user_id')

quality_df = quality_ex + ex_df
quality_df.fillna(0, inplace=True)
quality_df = quality_df.reset_index()
quality_df
```

```python
quality_df.drop(columns=['quality_3','quality_4'], inplace=True)
quality_df
```

```python
train_X = pd.concat([quality_df, error_df], axis=1)
train_X
```

```python
problem = np.zeros(15000)
# error와 동일한 방법으로 person_idx - 10000 위치에 
# person_idx의 problem이 한 번이라도 발생했다면 1
# 없다면 0
problem[train_problem.user_id.unique()-10000] = 1 
problem.shape
```

```python
problem_y = pd.DataFrame(problem, columns=['y'])
problem_y = problem_y.reset_index()
problem_y.rename(columns={'index':'user_id'}, inplace=True)
problem_y.user_id = list(map(lambda x:x+10000,problem_y.user_id))
problem_y
```

```python
train_df = pd.merge(train_X,problem_y,how='inner',on='user_id')
train_df
```

```python
X_train = train_df[train_df.columns[1:-1]].values
y_train = train_df[train_df.columns[-1]]
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier().fit(X_train,y_train)
```

```python
rf.score(X_train,y_train)
```

```python
# Train
#-------------------------------------------------------------------------------------
# validation auc score를 확인하기 위해 정의
def f_pr_auc(probas_pred, y_true):
    labels=y_true.get_label()
    p, r, _ = precision_recall_curve(labels, probas_pred)
    score=auc(r,p) 
    return "pr_auc", score, True
#-------------------------------------------------------------------------------------
models     = []
recalls    = []
precisions = []
auc_scores = []
threshold = 0.5
# 파라미터 설정
params = {
            'boosting_type' : 'gbdt',
            'objective'     : 'binary',
            'metric'        : 'auc',
            'seed': 1015
            }
```

```python
#-------------------------------------------------------------------------------------
# 5 Kfold cross validation
k_fold = KFold(n_splits=5, shuffle=True, random_state=42)
for train_idx, val_idx in k_fold.split(X_train):

    # split train, validation set
    X = X_train[train_idx]
    y = y_train[train_idx]
    valid_x = X_train[val_idx]
    valid_y = y_train[val_idx]

    d_train= lgb.Dataset(X, y)
    d_val  = lgb.Dataset(valid_x, valid_y)
    
    #run traning
    model = lgb.train(
                        params,
                        train_set       = d_train,
                        num_boost_round = 1000,
                        valid_sets      = d_val,
                        feval           = f_pr_auc,
                        verbose_eval    = 20, 
                        early_stopping_rounds = 3
                       )
    
    # cal valid prediction
    valid_prob = model.predict(valid_x)
    valid_pred = np.where(valid_prob > threshold, 1, 0)
    
    # cal scores
    recall    = recall_score(    valid_y, valid_pred)
    precision = precision_score( valid_y, valid_pred)
    auc_score = roc_auc_score(   valid_y, valid_prob)

    # append scores
    models.append(model)
    recalls.append(recall)
    precisions.append(precision)
    auc_scores.append(auc_score)

    print('==========================================================')
```

```python
print(np.mean(auc_scores))
```
