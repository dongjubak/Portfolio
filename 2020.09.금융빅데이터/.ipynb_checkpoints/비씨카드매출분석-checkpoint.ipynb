```python
len(temp201904['가맹점업종코드'].unique())
```

```python
onjob201904 = []

for i in range(len(temp201904['가맹점업종코드'].unique())):
    job = temp201904[(temp201904.가맹점업종코드 == temp201904['가맹점업종코드'].unique()[i])]
    onjob201904.append(job)
```

```python
# IDEA!) 유동인구 분포 수에 따른 업종의 매출금액을 예상해주는 모델 생성. (새로 사업,창업하려는 사람들에게 필요한 정보 제공 및 컨설팅 용도로.)

# ( 성별 / 연령별 / 업종별 ) 에 따른 매출금액을 파악 및 학습한다.
# + 가능하다면 가맹점신우편번호와 고객신우편번호도 살짝 보고.
# **) 단, 일자별에 따른 정보(ex. 주말, 공휴일에 더 많이 팔리고 등등)은 생략하자. << 일수가 16일까지밖에 없다. 달도 2달밖에 없고.

# 유동인구 데이터를 분석하고ㅡ 한 지역의 유동인구의 분포(ex.20여자 몇명, 30남자 몇명)를 파악한다.
# 유동인구 분포 수에 따라서 업종의 매출금액을 예상해주는 모델을 만든다.
```
