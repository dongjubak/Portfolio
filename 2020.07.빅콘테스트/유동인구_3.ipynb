```python
def make_plot_HDONG():
    fig, axes = plt.subplots(nrows=10,ncols=7)
    fig.set_size_inches(12, 30)
    idx_HD = 0
    x = 0
    y = 0
    day_of_month = 28
    HDONG = []
    index2 = 0

    for index in range(0,1932,day_of_month):
        HDONG.append(train_df[index:index+day_of_month])

    while idx_HD < 69:
        for x in range(10):
            for y in range(7):
                axes[x][y].plot(HDONG[idx_HD][HDONG[idx_HD].columns[1]], HDONG[idx_HD][HDONG[idx_HD].columns[34]]) # 남 나이대 합
                axes[x][y].plot(HDONG[idx_HD][HDONG[idx_HD].columns[1]], HDONG[idx_HD][HDONG[idx_HD].columns[35]]) # 여 나이대 합
                axes[x][y].plot(HDONG[idx_HD][HDONG[idx_HD].columns[1]], HDONG[idx_HD][HDONG[idx_HD].columns[36]]) # 총 합
                axes[x][y].set_title(HDONG[idx_HD].HDONG_NM.values[0])
                idx_HD += 1
                
    mpl.rc('font', family='NanumGothic')
    plt.show()

make_plot_HDONG()

# list index out of range 라고 뜨면서 된다.
```
