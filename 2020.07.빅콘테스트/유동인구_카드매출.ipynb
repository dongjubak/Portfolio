```python
x_data = S_J_sum[S_J_sum.columns[1:13]]
y_data = S_J_sum['총이용금액']
```

```python
print(x_data.shape, y_data.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.33, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn import linear_model

sk_lr = linear_model.LinearRegression(normalize=False)
sk_lr.fit(X_train, y_train)

sk_lr.intercept_ , sk_lr.coef_
```

```python
print("선형회귀 훈련 세트 점수: {}".format(sk_lr.score(X_train, y_train)))
print("선형회귀 테스트 세트 점수: {}".format(sk_lr.score(X_test, y_test)))
```

```python
from sklearn.linear_model import Lasso, Ridge
```

```python
ridge = Ridge(fit_intercept=True, alpha=1)
ridge.fit(X_train,y_train)
#lasso = Lasso(fit_intercept=True, alpha=0.5)
```

```python
ridge.intercept_ , ridge.coef_
```

```python
print("릿지 훈련 세트 점수: {}".format(ridge.score(X_train, y_train)))
print("릿지 테스트 세트 점수: {}".format(ridge.score(X_test, y_test)))
```

```python
lasso = Lasso(fit_intercept=True, alpha=1, max_iter = 100000)
lasso.fit(X_train,y_train)
```

```python
lasso.intercept_ , lasso.coef_
```

```python
print("랏소 훈련 세트 점수: {}".format(lasso.score(X_train, y_train)))
print("랏소 테스트 세트 점수: {}".format(lasso.score(X_test, y_test)))
print("사용한 특성의 개수: {}".format(np.sum(lasso.coef_!=0)))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF = RandomForestClassifier(n_estimators=10,random_state=10).fit(X_train,y_train)
print("RandomForest")
print("훈련 세트 정확도: {}".format(RF.score(X_train,y_train)))
print("테스트 세트 정확도: {}".format(RF.score(X_test,y_test)))
```
