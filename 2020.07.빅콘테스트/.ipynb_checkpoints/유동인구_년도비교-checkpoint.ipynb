```python
plt.rc('font', family='NanumGothic')
fig = plt.figure(figsize=(22,14))
fig.suptitle('2월~5월까지, 일자별 총 유동인구 수', fontsize=20)
a = fig.add_subplot(2,2,1)

# 2019년은 2~5월 총 일수 120일. 2020년은 121일.
day_2019 = [i for i in range(120)]
day_2019 = np.array(day_2019)
day_2020 = [i for i in range(121)]
day_2020 = np.array(day_2020)

a.plot(day_2019, Flow_Age_2019[Flow_Age_2019.columns[36]].values[0:120], label='2019', color='blue')
a.plot(day_2020, Flow_Age_2020[Flow_Age_2020.columns[36]].values[0:121], label='2020', color='red')
a.legend()
plt.show()
```

```python
plt.rc('font', family='NanumGothic')
fig = plt.figure(figsize=(22,30))
a = fig.add_subplot(2,2,1)

# 2019년은 2~5월 총 일수 120일. 2020년은 121일.
day_2019 = [i for i in range(120)]
day_2019 = np.array(day_2019)
day_2020 = [i for i in range(121)]
day_2020 = np.array(day_2020)

a.barh(day_2019, Flow_Age_2019[Flow_Age_2019.columns[36]].values[0:120], label='2019')
a.barh(day_2020, -Flow_Age_2020[Flow_Age_2020.columns[36]].values[0:121], label='2020')
a.legend()
plt.show()
```
